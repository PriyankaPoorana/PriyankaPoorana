```python
import re
import phonenumbers
## email contains name, domain name and extension
email_regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'

##mobile number starts with +65 or +91 and contains numbers for 10 digits
mobile_regex = r'\b(65|91)[0-9]{10}\b'

def check_email(email):
    if(re.fullmatch(email_regex,str(email))):
        print('matching')
    else:
        print('not matching')
def check_mobile(mobile):
    if(re.fullmatch(mobile_regex,str(mobile))):
        print('matching')
    else:
        print('not matching')
    
if __name__ == '__main__':
    email=input('enter email')
    mobile=input('enter contact number')
    check_email(email)
    check_mobile(mobile)
    my_number = phonenumbers.parse(str(mobile))
    print(phonenumbers.is_possible_number(my_number))
```
